```python
from deeppavlov import build_model

model = build_model('squad_ru_bert', download=True)
```

```python
# парсером обрабатываем страницу с документацией

import requests
from bs4 import BeautifulSoup


def get_text(url):
    rs = requests.get(url)
    root = BeautifulSoup(rs.content, 'html.parser')
    article = root.select_one('article')

    return article.text


url = 'https://yandex.ru/dev/market/partner-dsbs/doc/dg/reference/post-campaigns-id-orders-status-update.html'
text = get_text(url)

print(text[:100])  # Первые 100 символов из строки
```

```python
%time model([text], ['Что означает DELIVERY?'])
```

```python
%time model([text], ['статус если заказ оформлен, но еще не оплачен?'])
```

```python
%time model([text], ['Когда создается статус DELIVERY?'])
```

```python
%time model([text], ['Способ возможного решения Order id orderId is in terminal status?'])
```
